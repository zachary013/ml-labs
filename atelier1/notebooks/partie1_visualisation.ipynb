```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Charger les jeux de données
salary_df = pd.read_csv('../data/Salary_Data.csv')
insurance_df = pd.read_csv('../data/insurance.csv')
```

```python
# 1. Exploration des données
print("Données du jeu Expérience / Salaire :")
print(salary_df.head())
print(salary_df.info())
```

```python
#Resume du jeu Assurance :
print(insurance_df.head())
print(insurance_df.info())
```

```python
# 2. Résumé statistique
print("Résumé statistique du jeu Expérience / Salaire :")
print(salary_df.describe())
```

```python

# Résumé statistique du jeu Assurance
print(insurance_df.describe())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(salary_df['YearsExperience'], salary_df['Salary'])
plt.xlabel('Années d expérience')
plt.ylabel('Salaire')
plt.title('Nuage de points Expérience / Salaire')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
pd.plotting.scatter_matrix(insurance_df, figsize=(10, 10))
plt.suptitle('Matrice de dispersion du jeu Assurance')
plt.show()
```
